### Bloque Preprocesamiento
Matriz de trade-off
| Caracteristica          | The Simple Moving Average | The Exponential Moving average |
|-------------------------|---------------------------|--------------------------------|
| Ajuste                  | +                         | +                              |
| Perdida de datos        | -                         | +                              |
| Costo de implementacion | +                         | -                              |
| Complejidad de calculos | +                         | -                              |
| Outlier                 | -                         | +                              |
| Ajuste a cambios        | -                         | +                              |
| Total                   | 3                         | 4                       

Dado el costo de implementacion , las caracteristicas de funcionamiento, el hardware reducido, la necesidad de alta velocidad de procesamiento y el menor uso de memoria se determina que el metodo "The Simple Moving Average" es el mas factible a implementar; A pesar de tener menor cantidad de puntos positivos dada la aplicacion este metodo se adapta mejor.       |

```python
import numpy as np
from scipy import stats
from scipy.stats import t
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
data = pd.read_excel('Dataset.xlsx')
tiempo=data['Tiempo(s)']
ppm=data['BPM']
```

Aplicacion del filtrado:


```python
ppm_filtrado=ppm.rolling(5, min_periods=1).mean()
##ppm_filtrado2=ppm.ewm(alpha=0.3, adjust=False).mean()

plt.figure()
plt.plot(tiempo, ppm, color='blue', label='Data 1')
plt.plot(tiempo, ppm_filtrado, color='red', label='Data 2')
##plt.plot(tiempo, ppm_filtrado2, color='green', label='Data 2')
plt.xlabel('Tiempo[s]')
plt.ylabel('ppm')
plt.title('Comparison of Two Datasets')
plt.legend(['Original','Filtrado SMA','Filtrado EMA'])
plt.grid(True)
plt.show()
```

En le grafico se observa que la curva de ajuste con ese metodo es apegada a la grafica original, pero eliminando el ruido lo cual es una ventaja al hacer el analisis de datos, el filtro es efectivo ya que conserva las caracteristicas de la onda pero eliminando los picos y sin atenuar de manera significativa la amplitud ni retrasar la onda.
